# Pandas Project booking information for a hotel

```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

#Importing DataSet and reading the csv file

```python
hotels = pd.read_csv('hotel_booking_data.csv')
```

```python
#Head: used to display 1st 5 rows
hotels.head()
```

```python
#The info(): method prints information about the DataFrame. The information contains the number of columns,
#column labels, column data types, memory usage, range index, and the number of cells in each column
hotels.info()
```

```python
#df.isnull().sum(): get the count of null values in a column. 
hotels.isnull().sum()
```

```python
#The describe() method returns description of the data in the DataFrame.
#If the DataFrame contains numerical data, the description contains these information for each column
hotels.describe()
```

```python
#Drop: will drop the existing column that u need.
hotels=hotels.drop('company',axis=1)
```

#What are the top 5 most common country codes in the dataset?

```python
#Value_Counts:Return a Series containing counts of unique values.
hotels['country'].value_counts()[:5]
```

#What is the name of the person who paid the highest ADR (average daily rate)? How much was their ADR?

```python
#sorting: predicting the placement positions of unsorted data elements in a sorted sequence based on machine learning models 
#The iloc() function in python is defined in the Pandas module that helps us to select a specific row or column from the data set. 
hotels.sort_values('adr',ascending=False)[['adr','name']].iloc[0]
```

#What is the average (mean) number of nights for a stay across the entire data set? Feel free to round this to 2 decimal points.

```python
hotels['total_stay_nights']=hotels['stays_in_week_nights']+hotels['stays_in_weekend_nights']
round(hotels['total_stay_nights'].mean(),2) 

#The round() function returns a floating point number that is a rounded version of the specified number, 
#with the specified number of decimals.
```

#.What is the average total cost for a stay in the dataset?

```python
hotels['total_spend'] = hotels['adr']*hotels['total_stay_nights']
round(hotels['total_spend'].mean(),2)
```

#What are the names and emails of people who made 5 "Special Requests"?

```python
hotels[hotels['total_of_special_requests']==5][['name','email']]
```

#What percentage of hotel stays were classified as "repeat guests"?

```python
#The round() function returns a floating point number that is a rounded version of the specified number, 
#with the specified number of decimals.
round(100*sum(hotels['is_repeated_guest']==1) / len(hotels),2)
```

#What are the top 5 most common last name in the dataset?

```python
#Lambda: when you need to access several services or do custom processing
hotels['name'].apply(lambda name : name.split()[-1]).value_counts()[:5]
```

#What are the names of the people who had booked the most number children and babies for their stay?

```python
#sorting: predicting the placement positions of unsorted data elements in a sorted sequence based on machine learning models 
hotels['total_kids'] = hotels['babies'] + hotels['children']
hotels.sort_values('total_kids', ascending=False)[['name', 'total_kids']]
```

#What are the top 3 most common area code in the phone numbers? (Area code is first 3 digits

```python
#Lambda: when you need to access several services or do custom processing
hotels['phone-number'].apply(lambda num : num[:3]).value_counts()
```
